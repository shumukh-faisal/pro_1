Store System - README

## 📌 Project Description
This project is a simple Store Management System written in Python.  
It allows the user to manage products by:

- Add product
- Delete product
- Update product
- Display products
- Save products to file
- Load products from file

The system also uses:
- (OOP)
- Inheritance
- Encapsulation
- File Handling


##Classes Used

### 1. Product Class
This is the main class that contains:

- Product name
- Product price (private)
- Product quantity

Methods:
- get_price() → return price  
- set_price() → update price safely


```python
class product:
    def __init__(self, name_product, price, quantity):
        self.name_product = name_product
        self.__price = price
        self.quantity = quantity

    def get_price(self):
        return self.__price

    def set_price(self, price):
        if price > 0:
            self.__price = price
```

2. Stor Class
This class inherits from product class and adds:

- Product type

Example:
- Food
- Electronics
- Clothes


```python
class stor(product):
    def __init__(self, name_product, price, quantity, pro_type):
        super().__init__(name_product, price, quantity)
        self.pro_type = pro_type
```

# Add Product
Adds a new product to the store

### Delete Product
Deletes product using product name

### Update Product
Updates:
- Price
- Quantity

### Display Products
Shows all products in the system

```python
products = []

def add():
    name_product = input("enter product name: ")
    price = float(input("enter price: "))
    quantity = int(input("enter quantity: "))
    pro_type = input("enter type of product: ")

    product = stor(name_product, price, quantity, pro_type)
    products.append(product)

    print("add successfully")

def delete():
    name_product = input("enter product name to delete: ")

    for p in products:
        if p.name_product == name_product:
            products.remove(p)
            print("Deleted successfully")
            return

    print("not found")

def update():
    name_product = input("enter product name to update: ")

    for p in products:
        if p.name_product == name_product:
            price = float(input("enter new price: "))
            quantity = int(input("enter new quantity: "))

            p.set_price(price)
            p.quantity = quantity

            print("Updated successfully")
            return

    print("not found")

def display():
    for p in products:
        print("Product Name:", p.name_product)
        print("price:", p.get_price())
        print("quantity:", p.quantity)
        print("type:", p.pro_type)
```

###Save to File
Saves data into:
products.txt

### Load from File
Loads saved data when program starts


```python
def save_to_file():
    file = open("products.txt", "w")

    for p in products:
        file.write(
            p.name_product + "," +
            str(p.get_price()) + "," +
            str(p.quantity) + "," +
            p.pro_type + "\n"
        )

    file.close()

def load_from_file():
    try:
        file = open("products.txt", "r")

        for line in file:
            name, price, quantity, category = line.strip().split(",")

            product = stor(
                name,
                float(price),
                int(quantity),
                category
            )

            products.append(product)

        file.close()

    except:
        pass
```

##  Menu Options

1. Add Product: enter name, price, and quantity  
2. Delete Product: remove a product  
3. Update Product: edit product details  
4. Display Products: show all products  
5. Exit: save data and close the program

```python
def main():
    load_from_file()

    while True:
        print("\nStore System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Display Products")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add()

        elif choice == "2":
            delete()

        elif choice == "3":
            update()

        elif choice == "4":
            display()

        elif choice == "5":
            save_to_file()
            print("Goodbye")
            break

        else:
            print("Invalid choice")

main()
```
